#### Question 1.1

```python
n = 240
av = [0, 20, 0, 0, 0, 0]
iv = [30, 45, 60, 40, 40, 40]
ones = [15, 20, 25, 15, 20, 25]

TC = {}
for i in range(0, 6):
    # get values from vectors
    arrival = av[i]
    increment = iv[i]
    active = ones[i]
    
    period = increment
    count = 0
    periods = []
    while period < (n - arrival - increment):
        period = arrival + increment * count
        periods.append(period)
        count+=1
    
    #initialize temporal source as vector of 0s
    tempSource = [0 for i in range(n+20)]
    
    for start in periods:
        count = 0
        activeCount = 0
        while count < increment:
            if activeCount < active:
                tempSource[start + count] = 1
                count+=1
                activeCount=activeCount+1
                
            else:
                count+=15
                activeCount = 0

    TC[i] = tempSource[:240]

TC = pd.DataFrame(TC)

scaler = StandardScaler()
TC = scaler.fit_transform(TC)
```

```python
def lasso_regression(rho, TC, X):
    step = 1 / (np.linalg.norm(np.dot(TC, TC.transpose()) * 1.1))
    thr = rho * X.shape[0] * step
    
    A_o = np.zeros((TC.shape[1], 1))
    A = np.zeros((TC.shape[1], 1))
    A_lr = np.zeros((TC.shape[1], X.shape[1]))
    
    for k in range(X.shape[1]):
        A = A_o + step * np.dot(TC.transpose(), (X[:, k] - np.dot(TC, A_o).transpose()).transpose())
        A = (1 / (1 + thr)) * (np.sign(A) * np.maximum(np.zeros((TC.shape[1], 1)), np.absolute(A) - thr))
        for i in range(10):
            A_o = A
            A = A_o + step * np.dot(TC.transpose(), (X[:, k] - np.dot(TC, A_o).transpose()).transpose())
            A = (1 / (1 + thr)) * (np.sign(A) * np.maximum(np.zeros((TC.shape[1], 1)), np.absolute(A) - thr))
        A_lr[:, k] = A.transpose()
    return A_lr
```

```python
MSE_data = []
for rho in range(0, 105, 5):
    MSE_avg = 0
    for i in range(0, 10):
        np.random.seed = i * rho
        gamma_T = np.random.normal(0, sqrt(0.25), size = (240, 6))
        
        gamma_S = np.random.normal(0, sqrt(0.015), size = (6, 441))
        
        X_new = np.matmul((TC + gamma_T), (tmpSM + gamma_S))

        A_LR = lasso_regression(rho/100, TC, X_new)
        D_LR = np.matmul(X_new, A_LR.transpose())
        MSE_avg += np.linalg.norm((X_new - np.matmul(D_LR, A_LR)), ord = 'fro') / (240 * 441) # N * V
    
    MSE_data.append([rho/100, MSE_avg/10])
        
MSE_data
```

```python
MSE_data = np.array(MSE_data).transpose()
```
